```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

***LOAD DATASET***

```
url ="https://raw.githubusercontent.com/puneet-goel/learn/main/linearX.csv"
X = pd.read_csv(url)

url="https://raw.githubusercontent.com/puneet-goel/learn/main/linearY.csv"
Y = pd.read_csv(url)

X=(X.values).reshape((99,1))
Y=(Y.values).reshape((99,))

print(X)
print(Y)
```

```
def error(X,Y,theta,bias):
  return np.sum( (Y -(np.dot(X,theta) + bias))**2 )/(2*X.shape[0])
  
  
```

```
def gradient_descent(X,Y,learning_rate,iterations):
   theta = np.zeros(X.shape[1])
   bias = 0.0
   n = X.shape[0] 
 
   for i in range(iterations):
     predicted = (np.dot(X,theta)+bias)-Y
     grad = (1/n)*np.dot(X.T,predicted)
     db = np.sum(predicted)*(1/n)
     print(error(X,Y,theta,bias))
     theta -= grad*learning_rate
     bias  -= db*learning_rate
   return theta ,bias
```

```
theta , bias = gradient_descent(X, Y , learning_rate = 0.019,iterations = 500)
print(theta,bias)
```
